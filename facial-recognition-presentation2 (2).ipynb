## More Research:

<font size="4">Given the vast benefits that come from facial recognition models, it is important to recognize the detrimental limitations that are reflected in the algorithms, particularly for identifying women and black individuals. Graduate student Joy Buolamwini touches on this issue in her TED talk where she coined the term “coded gaze,” describing algorithmic bias which results in, as she states, “exclusionary experiences and discriminatory practices.” Thus, as facial recognition technology continues to grow, society must work towards diminishing these biases and improving identification of all faces equally regardless of race or gender.</font> <br>
<br>
<font size="4">Being aware of this potential bias, we are going to include various demographics (gender and race) in our project to determine whether the model is biased or not.</font>
